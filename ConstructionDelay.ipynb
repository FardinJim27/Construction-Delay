### Outlier Detection

```python
#outlier detection & IQR
def findOutliers(df):
    outliers = []
    Q2 = df.median()
    Q1 = df.describe()[4]
    Q3 = df.describe()[6]
    
    # Q1 = dataset.quantile(0.25)
    # Q3 = dataset.quantile(0.75)
    
    IQR = Q3 - Q1
    lf = Q1 - 1.5*IQR # lf = lower fence
    uf = Q3 + 1.5*IQR # Uf = Upper fence
    for x in df:
        if x < lf or x > uf:
            outliers.append(x)
    return outliers
```
